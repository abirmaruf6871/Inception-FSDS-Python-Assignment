```python
import google.generativeai as genai
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
GEMINI_API_KEY=os.getenv("GEMINI_API_KEY")
```

```python
user_text="""
Give me the sentiment analysis of the following review:
"I absolutely loved the new cafe in town! 
The ambiance was cozy, the staff were friendly, 
and the coffee was top-notch. I will definitely be coming back here often."


"""

user_text2="""
write bangla for this following sentences:
"I absolutely loved the new cafe in town!
The ambiance was cozy, the staff were friendly,
and the coffee was top-notch. I will definitely be coming back here often."
"""

genai.configure(api_key=GEMINI_API_KEY)
model=genai.GenerativeModel("gemini-2.5-flash")
response=model.generate_content(user_text2)
results=response.text
print(results)
```
