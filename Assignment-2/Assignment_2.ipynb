**1. Email Validator (String Filtering + Loops)**

ðŸ’¼ Real-world use: Used in signup forms or data cleaning tasks.

Task:

Check if a given email is valid â€” must contain â€œ@â€ and end with â€œ.comâ€ or â€œ.orgâ€.

Example:
Input: hello@inceptionbd.com

Output: Valid Email

```python
email=str(input("Enter your Email: "))
#taking ending text in a lists
ending_text=['.com','.org']

#checking for @ availability
if "@" in email:
  #checking ending text
  for e in ending_text:
    if email.endswith(e):
      print("Valid Email")
      break

  else:
    print("Ending must be with '.com','.org'")

else:
  print("must contain '@' and valid ending.")
```

**2. Count Word Frequency in Customer Reviews**

ðŸ’¼ Real-world use: Analyzing reviews or comments in text analytics.
Task:

Given a review text, count how many times each word appears (ignore case).

Example:

Input: "This product is good and this service is excellent"

Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
review_text = str(input("Enter a review: ")).lower()
text_list = review_text.split()

unique_words = []

for w in text_list:
    if w not in unique_words:
        unique_words.append(w)

for u in unique_words:
    count = 0
    for w in text_list:
        if u == w:
            count += 1
    print(u, ":", count)
```

**3. Password Strength Checker**

ðŸ’¼ Real-world use: Used in user authentication systems.

Task:
Check if a password is strong â€” must have at least one uppercase, one lowercase, one digit, and one
special character.

Example:
Input: "Hello@123"

Output: Strong Password

```python
password=str(input("Enter your Password:"))
upper=0
lower=0
digit=0
special=0

#storing data in variable
for i in password:
  if i.isupper():
    upper+=1

  elif i.islower():
    lower+=1

  elif i.isdigit():
    digit+=1

  elif not i.isalnum():
    special+=1

#conditions for warning
if upper==0:
  print("Weak Password! Your password must have 1 uppercase character.")

if lower==0:
  print("Weak Password! Your password must have 1 lowercase character.")

if digit==0:
  print("Weak Password! Your password must have 1 digit.")

if special==0:
  print("Weak Password! Your password must have 1 special character.")

if upper > 0 and lower > 0 and digit > 0 and special > 0:
  print("Strong_password!")




```

**7. Detect Suspicious Words in Text (Moderation)**

ðŸ’¼ Real-world use: Used in chat moderation or email filtering.

Task:

If any â€œbannedâ€ words like ['spam', 'scam', 'fake'] are found in a text, print â€œWarningâ€.

Example:

Input: "This is a scam offer"

Output: Warning: Suspicious content detected

```python
#user input
user_input=str(input("Enter a Sentence:"))
#banned words lists
banned=['spam', 'scam', 'fake']

#logic for warning message
for b in banned:
  if b in user_input:
    print("Warning: Suspicious content detected")
    break

else:
  print("Thank You!")
```

**9. Count Hashtags and Mentions (Social Media Analytics)**

ðŸ’¼ Real-world use: Used in analyzing social media posts.

Task:

Count how many hashtags (#) and mentions (@) appear in a post.

Example:

Input: "Check out #AIWorkshop and follow @inceptionbd"

Output: Hashtags = 1, Mentions = 1

```python
#user input
post=str(input("Write a post here:"))
hashtag=0
mention=0

#logic for mention count
for p in post:
  if p=="@":
    mention+=1
print("Mentions = ",mention)

#logic for hashtag count
for p2 in post:
  if p2=="#":
    hashtag+=1
print("Hashtags = ",hashtag)
```

**10. Word Reversal for Data Encryption**

ðŸ’¼ Real-world use: Simple text obfuscation or playful encoding.

Task:

Reverse each word in a sentence while keeping word order the same.

Example:

Input: "AI is fun"

Output: "IA si nuf"

```python
#user input
user=str(input("Enter a sentence:"))

re_list=""
user_s=user.split()
for u in user_s:

  re_list+=u[::-1]+" "
print(re_list)
```
