**1. Inventory Price Calculator**

Real-world use: Used in e-commerce dashboards for price summaries.

Task:Write a function calculate_inventory_summary (prices) that:

 receives a list of product prices

 returns total_cost and average_cost

Example:

Input: [120, 250, 399, 150]

Output: (919, 229.75)

```python
#Function Defining
def calculate_inventory_summary(prices):
  """
  This function is for calculate the inventory summary like total cost and average
  Input:It takes a list of product prices
  Output:Returns total cost and average cost
  Date:20-11-2025

  """

  total_cost = 0
  count = 0

  for p in prices:
        total_cost += p
        count += 1

  return total_cost, total_cost / count


prices = list(map(float, input("Enter all prices separate by commas: ").split(',')))
result = calculate_inventory_summary(prices)
print(result)
```

**3. Product Category Counter**

Real-world use: Inventory/category-based analytics.

Task:

Write a function count_categories(products) that:

 takes a list of product names

 returns a dictionary counting each product type.


Example:

Input: ["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]

Output: {"Laptop": 3, "Smartphone": 2, "Monitor": 1, "Headphones": 1}

```python
def count_categories(products):
  """
  This function for count categories
  Input: it will take a list of product names
  Output: It will return dictionary conting each product type
  Date:22-11-2025
  """

  laptop=0
  smartphone=0
  monitor=0
  headphones=0

  for i in products:
    if "Laptop" == i:
      laptop+=1

    elif "Smartphone" == i:
      smartphone+=1

    elif "Monitor" == i:
      monitor+=1

    elif "Headphones" ==i:
      headphones+=1

    else:
      print(f'{i} is not valid category for this app')

  return {
      "Laptop":laptop,
      "Smartphone":smartphone,
      "Monitor":monitor,
      "Headphones":headphones
  }

products= ["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]

result=count_categories(products)
print(result)
```

**4. Temperature Analytics**

Real-world use: Health tech temperature screening systems.

Task:

Write a function analyze_temperatures(temp_list) that:

 separates high (> 38°C) and normal temps

 returns two lists: (high_list, normal_list)


Example:

Input: [36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]

Output:

High List (>38°C): [39.1, 40.5]

Normal List (<38°C): [36.5, 38.0, 37.2, 38.0, 37.9]

Full Output Tuple: ([39.1, 40.5], [36.5, 38.0, 37.2, 38.0, 37.9])

```python
def analyze_temparatures(temp_list):
  """
  This function analyze the temparature based on condition and stored data in two empty list
  Input: list of temparatuees
  Output: High list ,normal list and full output tuple
  Date:22-11-2025
  """
  high_list=[]
  normal_list=[]

  #storing values for high_list & normal_list
  for temp in temp_list:
    if temp >38:
      high_list.append(temp)
    else:
      normal_list.append(temp)

  print(f'High List (>38°C):{high_list}')
  print(f'Normal List (<38°C):{normal_list}')

  full=(high_list,normal_list)


  return f'Full Output Tuple:{full}'

temp_list=[36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]

result=analyze_temparatures(temp_list)
print(result)
```

**5. Menu Price Lookup**

Real-world use: Restaurant ordering apps.

Task:

Write a function get_price(menu, item) that:

 receives a dictionary menu and an item name

 returns the price if exists

 otherwise returns "Item not found"


Example 1: Item Found

 Input (Menu): {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}

 Input (Item): "Muffin"

 Output: 4.00

Example 2: Item Not Found

 Input (Menu): {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}

 Input (Item): "Donut"

 Output: "Item not found"

```python
def get_price(menu, item):

  """
  This function is for to check user input and show the actual price as output
  Input:Food item name
  Output: Show the price of selected food item
  Date:23-11-2025
  """
  if item in menu:
    return menu[item]
  else:
    return "Item not found in menu"


menu = {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}

user_input = input("""Here is the menu, Please select your food:
Coffee-----3.50 Taka
Tea--------2.75 Taka
Sandwich---8.50 Taka

""")

result = get_price(menu, user_input)
print(result)
```

**6. Student Grading System**

Real-world use: School result automation.

Task:

Write a function calculate_grade(score) that:

 returns "A+", "A", "A-", "F" based on numeric marks

Write another function grade_students(student_dict) to:

 take a dictionary of student:score

 return student:grade dictionary

Example:

Input (Scores Dictionary): {"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69}

Output (Grades Dictionary): {"Alice": "A+", "Bob": "A", "Charlie": "F", "David": "A-"}

```python
def calculate_grade(score):
  """
  This function is used to determine a student's grade based on their score.
  Input: representing the student's mark.
  Output: Returns the appropriate grade as A+, A, A-, B, C, D or F
  Date: 23-11-2025
  """
  if score>=80:
    return "A+"

  elif score>=70 and score<80:
    return "A"

  elif score>=65 and score<70:
    return "A-"

  elif score>=60 and score<65:
    return "B"

  elif score>=50 and score<60:
    return "C"

  elif score>=40 and score<50:
    return "D"

  elif score>=0 and score<33:
    return "F"

  else:
    return "Invalid Input"


def grade_students(student_dict):
  """
  This function is for students and assigns grades to each of them.
  Input: A dictionary containing student names as keys and scores.
  Output: Returns a new dictionary mapping each student's name to their calculated grade.
  Date: 23-11-2025
  """

  gradewise_dict={}
  for name,score in student_dict.items():
    grade=calculate_grade(score)

    gradewise_dict[name] = grade
  return gradewise_dict




student_dict={"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69}
result=grade_students(student_dict)
print(result)
```

**7. Discount Calculator**

Real-world use: Online store discount engine.

Task:

Write a function apply_discounts(products) where:

 products is a list of tuples: (name, price, discount)

 return a list of updated prices after discount

Example:

Input (List of Tuples): [("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)]

Output (List of Prices): [64.00, 22.50, 285.00]

```python
def apply_discounts(products):

  """
  This function is for applying discounts
  Input:Takes a list of tuples
  Output:Return a list after discount
  Date:23-11-2025

  """
  price_list=[]

  for i in products:
    price=i[1]-i[1]*i[2]/100
    price_list.append(price)

  return price_list

products=[("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)]
result=apply_discounts(products)
print(result)
```

**8. Hashtag Generator**

Real-world use: Social media automation tools.

Task:

Write a function generate_hashtags(sentence) that:

 converts every word into a hashtag

 returns a list of hashtags

Example:

Input: "AI Data Science"

Output: ["#ai", "#data", "#science"]

```python
def generate_hashtags(sentence):
  """
  This function generate hashtag for every words
  Input: Takes a sentence
  Output:Return sentnce with hashtag for every words\
  Date:23-11-2025
  """
  new_list=[]
  words=sentence.split()
  for i in words:
    new_word="#"+i.lower()
    new_list.append(new_word)

  return new_list


sentence=input("Enter your sentence:")
result=generate_hashtags(sentence)
print(result)
```

**10. Shopping Cart Bill Calculator**

Real-world use: Billing systems in e-commerce apps.

Task:

Write a function calculate_total(cart, prices) that:

 takes a shopping cart list

 takes a dictionary of item:price

 returns the total bill amount


Example:

Input (Cart): ["Milk", "Bread", "Milk", "Cheese", "Bread"]

Input (Price): {"Milk": 3.00, "Bread": 2.50, "Eggs": 4.00, "Cheese": 5.00}

Output (Total Bill Amount): 16.00

```python
def calculate_total(cart, prices):
  """
  This function is for to calculate the total price from carts item
  Input:Shopping cart list and dictionary of prices
  Output:Total bill amount
  Date:23-11-2025
  """
  total_bill=0
  for item in cart:
    if item in prices:
      total_bill+=prices[item]

  return total_bill

cart= ["Milk", "Bread", "Milk", "Cheese", "Bread"]
prices= {"Milk": 3.00, "Bread": 2.50, "Eggs": 4.00, "Cheese": 5.00}

calculate_total(cart, prices)
```
