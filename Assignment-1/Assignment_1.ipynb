#Assignment 1: Temperature Check

Problem:
Write a program that takes the current temperature as input and prints:
* "It's too cold!" if below 10°C
* "It's cool outside" if between 10°C and 25°C
* "It's hot outside!" if above 25°C

```python
temp=float(input("Please write the current temparature(C) here: "))
if(temp<10):
  print("It's too cold")

elif(temp>=10 and temp<=25):
  print("It's cool outside")

else:
  print("It's hot outside!")
```

#Assignment 2: ATM Withdrawal System
Problem:
Ask the user for their account balance and the amount they want to withdraw.
* If the withdrawal amount is greater than the balance → print "Insufficient
Balance"
* If the withdrawal amount is less than or equal to balance → print "Transaction
Successful" and show the remaining balance.

```python
balance=float(input("Your Current Account Balance: "))
withdrawal_amount=int(input("Amount you want to withdraw: "))

if(balance<withdrawal_amount):
  print("Insufficient Balance")

else:
  print("Transaction Successful")
  rem_balance=balance-withdrawal_amount
  print("Remaining Balance", rem_balance)
```

#Assignment 3: Student Grade Calculator
Problem:
Input marks (0–100) and print grade based on:
* 90–100 → A+
* 80–89 → A
* 70–79 → B
* 60–69 → C
* Below 60 → Fail

```python
mark=float(input("Enter the mark to check your grade:"))

if(mark>=90 and mark<=100):
  print("Your Grade: A+")

elif(mark>=80 and mark<90):
  print("Your Grade: A")

elif(mark>=70 and mark<80):
  print("Your Grade: B")

elif(mark>=60 and mark<70):
  print("Your Grade: C")

elif(mark<60 and mark>=0):
  print("Your Grade: Fail")

else:
  print("Invalid Input")
```

#Assignment 4: Bus Ticket Discount
Problem:
Ask the user for age and print ticket price:
* Below 5 → Free
* 5–18 → 50% Discount
* 60 or above → 30% Discount
* Others → Full Price

```python
ticket_price=450
age=int(input("Enter your age: "))

if(age<5):
  print("Your ticket is Free")
  main_price=ticket_price-ticket_price
  print("Ticket Price:", main_price)

elif(age>=5 and age<=18):
  print("You are getting 50% Discount")
  main_price=ticket_price-ticket_price*0.50
  print("Ticket Price:", main_price)


elif(age>=60):
  print("You are getting 30% Discount")
  main_price=ticket_price-ticket_price*0.30
  print("Ticket Price:", main_price)

else:
  print("No Discount Available!")
  print("Ticket Price:", ticket_price)
```

#Assignment 5: Restaurant Bill with Discount
Problem:
Take total bill amount from the user:
* If the bill is above 1000 → apply 10% discount
* Otherwise → no discount
Then print the final bill amount.

```python
bill=float(input("Enter the bill amount: "))

if(bill>1000):
  print("Congrats! You are getting 10% discount.")
  final_bill=bill-bill*0.10
  discount_amount=bill*0.10
  print("Discount Amount: ", discount_amount)
  print ("You have to pay:", int(final_bill), "Taka")

else:
  print("No Discount!")
  print("You have to pay:", bill, "Taka")
```

#Assignment 8: Leap Year Checker
Problem:
Take a year as input and check:
* If it is divisible by 400 → Leap Year
* Else if divisible by 4 but not by 100 → Leap Year
* Otherwise → Not a Leap Year

```python
year=int(input("Enter a Year:"))

if(year%400==0):
  print(year, "is Leap Year")

elif(year%4==0 and year%100!=0):
  print(year, "is Leap Year")

else:
  print("Not a Leap Year")
```

#Assignment 10: Electricity Bill Calculator
Problem:
Input number of units consumed and calculate bill:
* Up to 100 units → 5 BDT per unit
* 101–200 units → 7 BDT per unit
* Above 200 units → 10 BDT per unit
Print total payable amount.

```python
units=float(input("Enter number of units consumed:"))

if(units>=0 and units<=100):
  print("Unit price:5 BDT")
  bill=units*5
  print("Your Bill:",bill,"Taka")

elif(units>100 and units<=200):
  print("Unit price:7 BDT")
  bill=units*7
  print("Your Bill:",bill,"Taka")

else:
  print("Unit price:10 BDT")
  bill=units*10
  print("Your Bill:",bill,"Taka")
```
